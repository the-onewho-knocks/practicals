```python
import numpy as np
```

```python
p1 = np.array([1, -1, 1, -1])
p2 = np.array([-1, 1, -1, 1])
p3 = np.array([1, 1, -1, -1])
p4 = np.array([-1, -1, 1, 1])

patterns = [p1, p2, p3, p4]

print("Stored Patterns:")
for i, p in enumerate(patterns):
    print(f"P{i+1}: {p}")
```

```python
size = len(p1)
W = np.zeros((size, size))

for p in patterns:
    W += np.outer(p, p)

# Remove self-connections
np.fill_diagonal(W, 0)

print("\nWeight Matrix:\n", W)
```

```python
noisy_input = np.array([1, -1, -1, -1])

print("\nNoisy Input:")
print(noisy_input)
```

```python
def recall(x, W, steps=5):
    x = x.copy()
    for _ in range(steps):
        for i in range(len(x)):
            net = np.dot(W[i], x)
            x[i] = 1 if net >= 0 else -1
    return x
```

```python
recalled = recall(noisy_input, W)

print("\nRecalled Pattern:")
print(recalled)
```
