```python
import numpy as np
```

```python
class BackPropagationNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        # Initialize weights and biases randomly
        self.W1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.randn(hidden_size, output_size)
        self.b2 = np.zeros((1, output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, X):
        # Forward propagation
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = self.sigmoid(self.z2)
        return self.a2

    def backward(self, X, y, output, learning_rate):
        # Output layer error
        output_error = y - output
        output_delta = output_error * self.sigmoid_derivative(output)

        # Hidden layer error
        hidden_error = np.dot(output_delta, self.W2.T)
        hidden_delta = hidden_error * self.sigmoid_derivative(self.a1)

        # Update weights and biases
        self.W2 += np.dot(self.a1.T, output_delta) * learning_rate
        self.b2 += np.sum(output_delta, axis=0, keepdims=True) * learning_rate
        self.W1 += np.dot(X.T, hidden_delta) * learning_rate
        self.b1 += np.sum(hidden_delta, axis=0, keepdims=True) * learning_rate

    def train(self, X, y, epochs, learning_rate):
        for epoch in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output, learning_rate)

            if epoch % 1000 == 0:
                loss = np.mean(np.square(y - output))
                print(f"Epoch {epoch}, Loss: {loss}")

    def predict_binary(self, X):
        output = self.forward(X)
        return (output > 0.5).astype(int)
```

```python
# XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

np.random.seed(42)
```

```python
bpn = BackPropagationNetwork(input_size=2, hidden_size=4, output_size=1)

print("=== FORWARD PROPAGATION (BEFORE TRAINING) ===")
output = bpn.forward(X)
print(output)
```

```python
bpn.train(X, y, epochs=10000, learning_rate=0.1)
```

```python
print("=== FORWARD PROPAGATION (AFTER TRAINING) ===")
output = bpn.forward(X)
print(output)
```

```python
predictions = bpn.predict_binary(X)

print("Input\tExpected\tPredicted")
for i in range(len(X)):
    print(f"{X[i]}\t{y[i][0]}\t\t{predictions[i][0]}")
```
